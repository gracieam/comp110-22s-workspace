We continue by importing the helper functions from `data_utils`.

```python
from csv import DictReader
from tabulate import tabulate
from data_utils import read_csv_rows, head, columnar, select, count, super_filter
```

Next, ... (you take it from here and add additional code and markdown cells to read in the CSV file and process it as needed)

```python
SURVEY_DATA_CSV_FILE_PATH: str = "../../data/survey.csv"
```

We will use the `read_csv_rows` function to read the rows of the survey file and turn those rows into a table.

```python
data_rows: list[dict[str, str]] = read_csv_rows(SURVEY_DATA_CSV_FILE_PATH)
```

Next, we need to transform the table that is currently a list of rows into a table represented as a dctionary of columns.

```python
data_cols: dict[str, list[str]] = columnar(data_rows)

print(f"Column names: {data_cols.keys()}")
```

The `head` function will be used to gather a visualization of what our data looks like. This function will return to us a number of rows specified by us. `Tabulate` will also be used here to show the data in a chart-like format.

```python
data_cols_head: dict[str, list[str]]= head(data_cols, 5)
tabulate(data_cols_head, data_cols_head.keys(), "html")
```

Now that we have a dictionary of columns and are able to see the first N amount of values in those columns, we will be able to visualize the relevant data. The `select` function can be used to pull the relevant columns for our data collection.

```python
selected_data: dict[str, list[str]] = select(data_cols, ["prior_exp", "difficulty"])
tabulate(head(selected_data, 20), selected_data.keys(), "html")
```

The relevant data is now apparent. Now we will make a function named `filter` to pull the data from students with all experience levels and make an illlustrative comparison. This will give us a clearer picture of how percieved difficulty level aligns with amount of coding experience.

```python
filtered: dict[str, list[str]] = super_filter(selected_data, "prior_exp", "difficulty")
tabulate(filtered, filtered.keys(), "html")
```

Finally, `count` will be used to tabulate the prior experience responses and the repsondents corresponding percieved diffulty level of the class. Furthermore, the built-in sorted function will be used to put the keys of the dictionary in order so that they will be plotted correctly.

```python
no_xp_count: dict[str, int] = count(filtered["None to less than one month!"])
sorted_no_xp_count: dict[str, int] = dict(sorted(no_xp_count.items()))
print(f"No Experience Count: {sorted_no_xp_count}")


two_to_six_mo: dict[str, int] = count(filtered["2-6 months"])
sorted_two_to_six_mo: dict[str, int] = dict(sorted(two_to_six_mo.items()))
print(f"2-6 Month Count: {sorted_two_to_six_mo}")

seven_to_twelve_mo: dict[str, int] = count(filtered["7-12 months"])
seven_to_twelve_mo['7'] = 0
sorted_seven_to_twelve_mo: dict[str, int] = dict(sorted(seven_to_twelve_mo.items()))
print(f"7-12 Month Count: {sorted_seven_to_twelve_mo}")

one_to_two_yrs: dict[str, int] = count(filtered["1-2 years"])
one_to_two_yrs['6'] = 0
one_to_two_yrs['7'] = 0
sorted_one_to_two_yrs: dict[str, int] = dict(sorted(one_to_two_yrs.items()))
print(f"1-2 Years Count: {sorted_one_to_two_yrs}")

over_two_yrs: dict[str, int] = count(filtered["Over 2 years"]) 
over_two_yrs['6'] = 0
over_two_yrs['7'] = 0
sorted_over_two_yrs: dict[str, int] = dict(sorted(over_two_yrs.items()))
print(f"Over 2 Years Count: {sorted_over_two_yrs}")
```

Using `matplotlib`, a bar chart will be made to demonstrate the count data for difficulty levels. Each difficulty levels will have a bar with the corresponding category of experience.

```python
from matplotlib import pyplot as plt
import numpy as np

plt.title("Count of Each Reported Difficulty Level For Each Level of Experience")
plt.xlabel("Difficulty Level")
plt.ylabel("Count of Reported Difficulty Level")

no_xp_values: list[int] = list(sorted_no_xp_count.values())
two_to_six_mo_values: list[int] = list(sorted_two_to_six_mo.values())
seven_to_twelve_mo_values: list[int] = list(sorted_seven_to_twelve_mo.values())
one_to_two_years_values: list[int] = list(sorted_one_to_two_yrs.values())
over_two_yrs_values: list[int] = list(sorted_over_two_yrs.values())

plt.title("Count of Each Reported Difficulty Level For Each Level of Experience")

x = 7
ind = np.arange(x)
width = 0.18

bar_1 = plt.bar(ind, no_xp_values, width, label="No Experince")
bar_2 = plt.bar(ind + width, two_to_six_mo_values, width, label='2-6 Months Experience')
bar_3 = plt.bar(ind + 2 * width, seven_to_twelve_mo_values, width, label='7-12 Months Experience')
bar_4 = plt.bar(ind + 3 * width, one_to_two_years_values, width, label='1-2 Years Experience')
bar_5 = plt.bar(ind + 4 * width, over_two_yrs_values, width, label='Over 2 Years Months Experience')

plt.xticks(ind+width, ["1", "2", "3", "4", "5", "6", "7"])
plt.legend( (bar_1, bar_2, bar_3, bar_4, bar_5), ("No Experience", "1-2 Months", "7-12 Months", "1-2 Years", "Over 2 Years"))
```
